```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("traffic.csv")
print(df.head()) 
```

```python
plt.scatter(df.temp,df.weather_main)
plt.show()
```

```python
plt.scatter(df.temp,df.weather_main)
sns.set_style("dark")
plt.show()
```

```python
plt.scatter(df.temp,df.weather_main)
sns.set_style("whitegrid")
sns.set_context("notebook")  
plt.show()
```

```python
#Histogram
df.head()
```

```python
sns.displot(df['traffic_volume'],kde=False)
```

```python
sns.set_context('talk')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x='weather_main', y='traffic_volume', data=df, hue='weather_main', palette='PuRd', errorbar=None)
plt.xticks(rotation=45) 
plt.show()
```

```python
# Horizontal Barplot

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='temp', y='weather_main', data=df, palette='PuRd', errorbar=None)  # Swap x and y for horizontal bars
plt.show()  # No need for plt.legend()
```

```python
sns.set_context('paper')

sns.countplot(x='temp', hue='weather_main',data=df,palette='magma')
plt.title('Survivors')
plt.show()
```

```python
import seaborn as sns

sns.jointplot(x='temp', y='weather_main', data=df, kind='scatter')  # or kind='hist'
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="temp", y="weather_main", data=df)
plt.show()
```
